#### Regresión por mínimos cuadrados

$$y=ax+b$$
$$a = \frac{N\sum{x_iy_i}-\sum{x_i}\sum{y_i}}{N\sum{x_i^2}-{(\sum{x_i})^2}}$$
$$b=\frac{N\sum{x_i^2}\sum{y_i}-\sum{x_i}\sum{x_iy_i}}{N\sum{x_i^2}-{(\sum{x_i})^2}}$$
$$r = \frac{\bar{xy}-\bar{x} \cdot \bar{y}}{\sqrt{\bar{x^2} - (\bar{x})^2} \cdot \sqrt{\bar{y^2} - (\bar{y})^2}}$$

Función generada

```julia
function RegresionLineal(x, y)
    N = length(x) # Elementos totales 
    suma_x = 0 # Inicializadores 
    suma_y = 0
    suma_xy = 0
    suma_x2 = 0
    suma_y2 = 0
    for i=1:N # Acumuladores
        suma_x = suma_x + x[i]
        suma_y = suma_y + y[i]
        suma_xy = suma_xy + x[i]*y[i]
        suma_x2 = suma_x2 + x[i]^2
        suma_y2 = suma_y2 + y[i]^2
    end 
    # Coeficientes y = ax + b
    a = (N*suma_xy - suma_x*suma_y)/(N*suma_x2 - suma_x^2)
    b = (suma_x2*suma_y - suma_x*suma_xy)/(N*suma_x2 - suma_x^2)
    yn = []
    for i=1:N
        append!(yn, a*x[i]+b)
    end
    # Calculo del ajuste r
    x_media = suma_x/N
    y_media = suma_y/N
    xy_media = suma_xy/N
    x2_media = suma_x2/N
    y2_media = suma_y2/N
    r = (xy_media - x_media*y_media)/(√(x2_media - x_media^2)*√(y2_media - y_media^2))
    # Convirtiendo vector x a una matrix x
    scatter(x, y)
    plot(transpose(x), yn, "r", label = "Lineal")
    title("Ajuste lineal")
    legend()
    println("Ajuste lineal")
    println(" y = ", a, "x ", b, "\nCon r = ", r)
end
```

```julia
RegresionLineal(x,y)
```

#### Modelo exponencial

$$y = ax + b$$
$$ln (y) = a \cdot ln (x) + b$$
$$e^{ln(y)} = e^{ln(x^n)+b}$$
$$y = e^bx^a$$
$$y = 10^bx^a$$

Función generada

```julia
function RegresionExponencial(x, y)
    N = length(x)
    ln_y = []
    ln_x = []
    suma_x = 0
    suma_y = 0
    suma_xy = 0
    suma_x2 = 0
    suma_y2 = 0
    for i=1:N # Acumuladores
        append!(ln_x, log(x[i]))
        append!(ln_y, log(y[i]))
        suma_x = suma_x + ln_x[i]
        suma_y = suma_y + ln_y[i]
        suma_xy = suma_xy + ln_x[i]*ln_y[i]
        suma_x2 = suma_x2 + ln_x[i]^2
        suma_y2 = suma_y2 + ln_y[i]^2
    end 
    # Coeficientes y = ax + b
    a = (N*suma_xy - suma_x*suma_y)/(N*suma_x2 - suma_x^2)
    b = (suma_x2*suma_y - suma_x*suma_xy)/(N*suma_x2 - suma_x^2)
    yn = []
    for i = 1:N
        append!(yn, exp(b)*x[i]^a)
    end
    # Calculo del ajuste r
    x_media = suma_x/N
    y_media = suma_y/N
    xy_media = suma_xy/N
    x2_media = suma_x2/N
    y2_media = suma_y2/N
    r = (xy_media - x_media*y_media)/(√(x2_media - x_media^2)*√(y2_media - y_media^2))
    scatter(x, y)
    plot(transpose(x), yn, "g", label = "Exponencial")
    legend()
    title("Ajuste exponencial")
    println("Ajuste exponencial")
    println(" y = ", exp(b), "x^", a, "\nCon r = ", r)
end
```

```julia
RegresionExponencial(x,y)
```

#### Ajuste logaritmico

```julia
x = [2 4 6 8 10 12]
y = [1.4 3.5 8.0 7.3 9.2 10.0]
using PyPlot
function RegresionLog(x, y)
    N = length(x)
    ln_y = []
    ln_x = []
    suma_x = 0
    suma_y = 0
    suma_xy = 0
    suma_x2 = 0
    suma_y2 = 0
    for i=1:N # Acumuladores
        append!(ln_x, log10(x[i]))
        append!(ln_y, log10(y[i]))
        suma_x = suma_x + ln_x[i]
        suma_y = suma_y + ln_y[i]
        suma_xy = suma_xy + ln_x[i]*ln_y[i]
        suma_x2 = suma_x2 + ln_x[i]^2
        suma_y2 = suma_y2 + ln_y[i]^2
    end 
    # Coeficientes y = ax + b
    a = (N*suma_xy - suma_x*suma_y)/(N*suma_x2 - suma_x^2)
    b = (suma_x2*suma_y - suma_x*suma_xy)/(N*suma_x2 - suma_x^2)
    yn = []
    for i = 1:N
        append!(yn, 10^b*x[i]^a)
    end
    # Calculo del ajuste r
    x_media = suma_x/N
    y_media = suma_y/N
    xy_media = suma_xy/N
    x2_media = suma_x2/N
    y2_media = suma_y2/N
    r = (xy_media - x_media*y_media)/(√(x2_media - x_media^2)*√(y2_media - y_media^2))
    scatter(x, y)
    plot(transpose(x), yn, "g")
    # yscale("log")
    # xscale("log")
    title("Ajuste logaritmico")
    println("Ajuste logaritmico")
    println(" y = ", 10^b, " x^ ", a, ")\nCon r = ", r)
end

RegresionLog(x, y)
```

#### Modelo polinomico de segundo orden

$$y = a_0+ a_1x + a_2x^2 $$

```julia
using LinearAlgebra
```

```julia
x = [2 4 6 8 10 12]
y = [1.4 3.5 8.0 7.3 9.2 10.0]
function RegresionCuadratica(x, y)
    x = float(x)
    y = float(y)
    n = length(x)
    suma_x = 0
    suma_x2 = 0
    suma_y = 0
    suma_x3 = 0
    suma_xy = 0
    suma_x4 = 0
    suma_x2y = 0
    for i=1:n
        suma_x = suma_x + x[i]
        suma_x2 = suma_x2 + x[i]^2
        suma_y = suma_y + y[i]
        suma_x3 = suma_x3 + x[i]^3
        suma_xy = suma_xy + x[i]*y[i]
        suma_x4 = suma_x4 + x[i]^4
        suma_x2y = suma_x2y + y[i]*x[i]^2
    end
    # Resolución de matriz con regla de Cramer
    A = [n suma_x suma_x2; suma_x suma_x2 suma_x3; suma_x2 suma_x3 suma_x4]
    b = [suma_y; suma_xy; suma_x2y]
    tam_matriz = size(A)
    DetA = det(A)
    ai = [] # [a0, a1, a2]
    for i=1:tam_matriz[1]
        Apivot = copy(A)
        Apivot[:,i] = b
        DetApivot = det(Apivot)
        a = DetApivot/DetA
        append!(ai, a)
    end
    # Aproximacion a los resultados
    yn = []
    for i = 1:n
        append!(yn, ai[1] + ai[2]*x[i] + ai[3]*x[i]^2)
    end
    # Cálculo del coeficiente de correlacion
    media_y = suma_y/n
    suma_1 = 0
    suma_2 = 0
    for i=1:n
        suma_1 = suma_1 + (y[i] - media_y)^2
        suma_2 = suma_2 + (y[i] - ai[1] - ai[2]*x[i] - ai[3]*x[i]^2)^2
    end 
    r = (suma_1 - suma_2)/suma_1
    # Gráfico
    scatter(x, y)
    plot(transpose(x), yn, "m", label = "Cuadrático")
    legend()
    title("Ajuste polinomico de segundo orden")
    println("Ajuste cuadratico")
    println(" y = (", ai[1], ") + (", ai[2], ") x + (", ai[3], ") x^2\nCon r = ", r) 
end
RegresionCuadratica(x, y)
```

#### Comparación de modelos

```julia
x = [2 4 6 8 10 12]
y = [1.4 3.5 8.0 7.3 9.2 10.0]

RegresionLineal(x, y)
RegresionExponencial(x, y)
RegresionCuadratica(x, y)
scatter(x,y)
title("Comparacion de ajustes")
```

De todos los ajustes realizados a los datos, encontramos que el modelo exponencial es quien nos arroja un mejor indice de correlación $r$. Por lo tanto, podriamos inferir que este es el que debemos seleccionar al momento de querer describir de manera más exacta el comportamiento de nuestra información. Sin embargo, el hacer esto o no depende del análisis posterior que deseemos realizar pues dicho modelo resulta más complicado de manipular matematicamente en comparación con sus alternativas. Si nos guiamos de esto podemos observar que el indice presentado en el modelo lineal es muy semejante al modelo cuadratico, por lo tanto, el seleccionar el primero de ellos puede ser una buena opción debido a su sencillez.
