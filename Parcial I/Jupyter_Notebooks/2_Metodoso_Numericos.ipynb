#### Metodos cerrados

##### Bisección

Corresponde a los métodos cerrados, pues requiere que se establezca dos limites $x_i$ y $x_f$ donde se encuentre la raiz de la función $f(x)$
En general, si $f(x)$ es real y continua en el intervalo que va desde $x_i$ y $x_f$ y se cumple que $f(x_i)$ y $f(x_f)$ tienen signos opuestos, es decir: 
$$f(a)f(b) < 0$$
Se dice que existe al menos un raiz entre el intervalo establecido
La aproximación de la raiz se determina mediante el siguiente criterio: $$x_n = \frac{x_i + x_f}{2}$$
Con este valor, se calcula un nuevo valor de la función $f(x_n)$, cuyo resultado puede cambiar de signo. Con esto, se realiza el siguiente analisis: 
- Si $f(x_i)f(x_n) < 0$, la raiz se encuentra dentro del intervalo $[x_i, x_n]$, por lo tanto ahora el nuevo valor de $x_f$ sera $x_n$ $$x_f = x_n$$
- Si $f(x_i)f(x_n) > 0$, la raiz se encuentra dentro del intervalo $[x_n, x_f]$, por lo tanto ahora el nuevo valor de $x_i$ sera $x_n$ $$x_i = x_n$$
- Si $f(x_i)f(x_n) = 0$, la raiz es igual a $x_n$ y se termina el cálculo

```julia
a0 = -5 #Punto incial
b0 = 5 #Punto final
ϵ = 0.0001 # Tolerancia del error
i = 1 #Iteracion 
xr = 2 #Aproximacion incial 
xr_list = []
fxr_list = []
i_list = []

while ecuacion(xr) <= -ϵ || ecuacion(xr) >= ϵ
    #Ecuacion del metodo
    xr = (a0 + b0)/2
    #Descartar subintervalos
    if ecuacion(a0)*ecuacion(xr) < 0
        b0 = xr 
    end 
    if ecuacion(xr)*ecuacion(b0) < 0
        a0 = xr 
    end
    # Redefinicion del criterio de error
    append!(xr_list, xr)
    append!(fxr_list,ecuacion(xr))
    append!(i_list, i)
    println("Iteracion ",i," = ", xr)
    i += 1
end
plot(i_list, xr_list, "ob")
xlabel("Dominio de la sucesión")
ylabel("Raiz")
```

##### Regla falsa

$x_r = b - \frac{f(b)(a-b)}{f(a)-f(b)}$ 

```julia
a0 = 0 # Primer limite
b0 = 10 # Segundo limite
ϵ = 0.0001 #Tolerancia del error
xr = 2 # Primera aproximación de la raiz
xr_list = []
fxr_list = []
i_list = []

while ecuacion(xr) <= -ϵ || ecuacion(xr) >= ϵ
    # Criterio de siguiente aproximacion de la raiz
    xr = b0 - (ecuacion(b0)*(a0-b0))/(ecuacion(a0)-ecuacion(b0))
    # Descarte de intervalos 
    if ecuacion(a0)*ecuacion(xr) < 0
        b0 = xr
    end
    if ecuacion(xr)*ecuacion(b0) < 0
        a0 = xr
    end
    # Redefinicion del criterio de error
    append!(xr_list, xr)
    append!(fxr_list,ecuacion(xr))
    append!(i_list, i)
    println("Iteracion ",i," = ", xr)
    i += 1
end 

plot(i_list, xr_list, "or")
xlabel("Dominio de la sucesión")
ylabel("Raiz")
```

##### Newton - Raphson

$$x_n = x_{n-1}-\frac{f(x)}{f'(x)}$$

Para proponer un punto inicial exiten dos metodos guia: 
- Boots Tranping
- Multi-start

```julia
function ecuacion(x)
    y = x^2 + 2*x - 4
    return y
end
```

```julia
function derive(x)
    dy = 2*x + 2
    return dy
end
```

```julia
xn = 0 #Valor inicial 3, -3, 0
ϵ = 0.0001 #Tolerancia del error
i = 1 #Iteracion

while ecuacion(xn) <= -ϵ || ecuacion(xn) >= ϵ
    xn = xn - ecuacion(xn)/derive(xn)
    println("Iteracion ", i, " = ", xn, " donde fx = ", ecuacion(xn))
    i += 1
end

println("La raiz por Newton - Raphson es ", xn)
```
