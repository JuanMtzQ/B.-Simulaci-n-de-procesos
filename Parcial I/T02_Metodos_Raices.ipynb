##### Biseccion

```julia
function Biseccion(fx, a0, b0, ϵ)
    i = 1 #Iteracion 
    xr = 2 #Aproximacion inicial 
    xr_list = [] # Vector de aproximaciones
    fxr_list = [] # Vector de resultados de la funcion
    i_list = [] # Vector de iteraciones
    while f(xr) <= -ϵ || f(xr) >= ϵ
        #Ecuacion del metodo
        xr = (a0 + b0)/2
        #Descartar subintervalos
        if f(a0)*f(xr) < 0
            b0 = xr 
        end 
        if f(xr)*f(b0) < 0
            a0 = xr 
        end
        # Redefinicion del criterio de error
        append!(xr_list, xr)
        append!(fxr_list, f(xr))
        append!(i_list, i)
        println("Iteracion ",i," = ", xr)
        i += 1
    end
plot(i_list, xr_list, "ob")
title("Solución de raices mediante bisección")
xlabel("Iteraciones")
ylabel("Raiz")
end
```

```julia
Biseccion(f, 0, 5, 0.0001)
```

##### Regla falsa

```julia
function ReglaFalsa(f, a0, b0, ϵ)
    i = 1 #Iteracion 
    xr = 2 #Aproximacion inicial 
    xr_list = [] # Vector de aproximaciones
    fxr_list = [] # Vector de resultados de la funcion
    i_list = [] # Vector de iteraciones
    while f(xr) <= -ϵ || f(xr) >= ϵ
        #Ecuacion del metodo
        xr = b0 - (f(b0)*(a0-b0))/(f(a0)-f(b0))
        #Descartar subintervalos
        if f(a0)*f(xr) < 0
            b0 = xr 
        end 
        if f(xr)*f(b0) < 0
            a0 = xr 
        end
        # Redefinicion del criterio de error
        append!(xr_list, xr)
        append!(fxr_list, f(xr))
        append!(i_list, i)
        println("Iteracion ",i," = ", xr)
        i += 1
    end
plot(i_list, xr_list, "ob")
title("Solución de raices mediante regla falsa")
ylabel("Raiz")
xlabel("Iteraciones")
grid("True")
end
```

```julia
ReglaFalsa(f, 0, 5, 0.0001)
```

##### Punto fijo

Sabemos que la raiz de la función se dara en el momento que esta y su función despejada $g(x)$ intersecten entre si

```julia
f(x) = x^2 + 2*x - 4
g(x) = (4-2*x)/2
x = collect(-5:0.1:5)
plot(x, f.(x))
plot(x, g.(x))
grid("True")
```

```julia
f(x) = x^2 + 2*x - 4
g(x) = (4-2*x)/2

x0 = -5
i = 1 
ϵ = 0.001
while f(x0) <= -ϵ || f(x0) >= ϵ
    xr = g(x0)
    x0 = xr
    i = i + 1
    println("Iteracion ", i, " xr = ", x0)
end
```

##### Newton-Raphson

```julia
function NewtonRaphson(fx, dfx, x0, ϵ)
    i = 1 # Iteracion
    i_list = []# Vector de iteraciones
    xn_list = [] # Vector de aproximaciones
    fx_list = [] # Vector de resultados fx
    dfx_list = [] # Vector de resultado dfx
    while f(x0) <= -ϵ || f(x0) >= ϵ
        # Criterio de aproximacion
        xn = x0 - f(x0)/dfx(x0)
        println("Iteracion ", i, " = ", xn)
        # Historial de resultados
        append!(i_list, i)
        append!(xn_list, xn)
        append!(fx_list, f(xn))
        append!(dfx_list, dfx(xn))
        x0 = xn
        i += 1
    end
plot(i_list, xn_list, "ob")
title("Solución de raices mediante punto fijo")
ylabel("Raiz")
xlabel("Iteraciones")
grid("True")
end
```

```julia
NewtonRaphson(f, df, 0, 0.0001)
```

##### Secante

```julia
f(x) = x^2 + 2*x - 4

function Dif_Atras(f, x0, h)
    x = x0 - h
    fx_prima = (f(x0) - f(x))/h
    return fx_prima
end

```

```julia
Dif_Atras(f, 1, 0.01)
```

```julia
f(x) = x^2 + 2*x - 4

function Dif_Atras(f, x0, h)
    x = x0 - h
    fx_prima = (f(x0) - f(x))/h
    return fx_prima
end

function Secante(f, x0, h, ϵ)
    i = 1
    while f(x0) <= ϵ || f(x0) >= ϵ
        xr = x0 - f(x0)/Dif_Atras(f, x0, h)
        x0 = xr
        println("Iteracion ", i, " xr = ", xr)
        i += 1
    end
end
```
